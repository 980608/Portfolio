```python
%%writefile dataset.yaml
path: /content/drive/MyDrive/yo11_data/

train: train/images
val: valid/images
test: test/images

names:
  0: Crack
  1: OtherDamage
  2: Pothole
```

```python
import os
print(len(os.listdir('/content/drive/MyDrive/yo11_data/train/images')))
print(len(os.listdir('/content/drive/MyDrive/yo11_data/valid/images')))
print(len(os.listdir('/content/drive/MyDrive/yo11_data/test/images')))
```

```python
model = YOLO("yolo11n.pt")

results = model.train(
    data="dataset.yaml",

    epochs=40,
    imgsz=640,
    batch=16,

    optimizer="AdamW",
    lr0=0.001,

    workers=2,
    patience=10,

    project="/content/drive/MyDrive/yolov11n_3class",
    name="YOLO11n_5class",

    save=True,
    plots=True
)
```
